```julia
####回溯线性搜索法
function inexact_method(f,g,ϕ0,dϕ0,xk,dk;τ=0.5,ϵ=0.5,ζ=2)
    α=1
    dα=dk'*g((xk.+α.*dk)...)##dk表示梯度/对α求导
    while (dα<0)
        α=ζ*α
        dα=dk'*g((xk.+α.*dk)...)
    end
    ϕα=f((xk.+α.*dk)...)
    while (ϕα>ϕ0+ϵ*α*dϕ0)
        α=τ*α
        ϕα=f((xk.+α.*dk)...)
    end
    return α
end
```

```julia
function gradient_descent(f,g,x0;ϵg=0.0001,ϵx=0.0001,ϵf=0.0001,maxIterations=128verbose=true)
    x1 = x0
    d = -g(x1...)####向量的形式
    ng=norm(d)
    f1=f(x1...)
    gx=g(x1...)
    i=1
    α=inexact_method(f,g,f1,-gx'*gx,x1,d)
    println("step",i,"\t$α",α)
    x2=x1+α.*d
    f2=f(x2...)
    nx=α*ng
    
    println("\tx=",x2,"\td=",d,"\tf=",f2)
    while (ng>ϵg || nx>ϵx || abs(f2-f1)>ϵf || i<maxIterations)
        x1=x2
        d=-g(x1...)
        gx=g(x1...)
        ng=norm(d)
        f1=f2
        α=inexact_method(f,g,f1,-gx'*gx,x1,d)
        x2=x1+α.*d
        f2=f(x2...)
        nx=α*ng
        i+=1
        if verbose
            println("step",i,"\tα=",α,"\tx=",x2,"\td=",d,"\tf=",f2)
        end
    end
    if i>=maxIterations
        println("WARNING:iterations exceeds",maxIterations,"steps before converges")
    else
        println("LOG:iteration converges after ",i,"step")
    end
    return x2,f2
end
```
