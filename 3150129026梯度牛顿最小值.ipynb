梯度下降法

```julia
function inexact_method(f,g,φφ0,φφd0,xk,dk;ττ=0.5,εε=0.5,ζζ=2)
           αα=1
           dαα=dk*g((xk + αα * dk)'...)'
           while (dαα<0)
               αα=ζζ*αα
               dαα=dk*g((xk + αα* dk)'...)'
           end
           φφαα=f((xk+ αα* dk)'...)
           while (φφαα>φφ0+εε+φφd0*αα)
               αα=ττ*αα
               φφαα=f((xk + αα* dk)'...)
           end
           return αα
       end
```

```julia
function gradient_descent(f,g,x0;
               εεg=0.0001,εεx=0.0001,εεf=0.0001,maxIterations=2000,verboss=true)
           x1=x0
           d=-g(x1...)
           f1=f(x1...)
           ng= norm(d)
           # Inexact method for |alpha
           alpha = inexact_method(f,g,f1,-ng*ng,x1,d)
           nx=alpha*ng
           x2=x1+alpha .* d
           f2=f(x2...)
           ni=1
           while((ng>εεg||nx>εεx||abs(f2-f1)>εεf)&&ni<maxIterations)
               x1=x2
               d=-g(x1...)
               f1=f2
               ng=norm(d)
               alpha = inexact_method(f,g,f1,-ng*ng,x1,d)
               x2=x1+alpha .* d
               f2=f(x2...)
               nx=alpha*ng
               ni+=1
               if verboss
                   println("step",ni," x ",x2," f ",f2)
                   println("        d  ",d,"   alpha ",alpha)
               end
           end
           if ni >=maxIterations
               println("WARNING:interations exceeds ",maxIterations," steps befo
re converges.")
           else
               println("LOG;interations converges after ",ni,"steps.")
           end
           return x2,f2
       end
```
