```julia
function inexact_method( f ,g ,ϕ0 ,ϕd0 ,xk , dk ;τ=0.5, ϵ=0.5, ζ=2)##回溯搜索法
    a=1
    da=dk' * g((xk .+ a .* dk)...)    ###  对f(xk .+ a .* dk)求导
    while ( da<0)
        a=ζ*a
        da=dk'* g((xk .+a .* dk)...)
    end
    ϕa=f((xk .+a .* dk)...)
    while ( ϕa > ϕ0 + ϵ * a * ϕd0 )
        a=τ*a
        ϕa=f((xk .+a .* dk)...)
    end
    return a 
end
```

```julia
function gradient_descent(f,g,x0;
        eg=0.001,ex=0.001,ef=0.001,maxIterations=128,verbose=true)###梯度下降法
    x1 = x0
    d  = -g(x1...)
    f1 = f(x1...)
    g1 = g(x1...)
    println("Step " ,"  x=",x1,"  f=",f1,"  g=",g1)
    ###Inexact method for a
    a  = inexact_method(f,g,f1,-g1'*g1,x1,d)
    ng = norm(d)   ##  梯度的范数<ϵ。
    nx = a*ng      ##  a*d的范数<ϵ。
    ni = 1
    x2 = x1+a.*d
    f2 = f(x2...)
    g2 = g(x2...)
    println("Step",ni,"  x=",x2,"  f=",f2,"  g=",g2)
    println("        d=",d, " a=",a )
    while (ng>eg || nx>ex || abs(f2-f1)>ef)&& ni < maxIterations#!(ng<eg && nx<ex && abs(f2-f1)<ef) && ni < maxIterations
        x1 = x2
        d  = -g(x1...)
        f1 = f2
        g1 = g(x1...)
        a  = inexact_method(f,g,f1,-g1'*g1,x1,d)
        ng = norm(d)
        nx = a*ng
        ni += 1
        x2 = x1+a .*d
        f2 = f(x2...)
        g2  = g(x2...)
        if verbose
            println("Step",ni,"  x=",x2,"  f=",f2,"  g=",g2)
            println("        d=",d, " a=",a )
        end
        if ni ==maxIterations
            println("WARNING: Interations exceeds ",maxIterations," steps")
        #else 
            #println(" LOG:  Iterations converges after ",ni," steps")
        end
    end
    return(x2,f2)
end
```

```julia
function Newton(f,g,h,x0;
        ϵg=0.001,ϵx=0.001,ϵf=0.01,ϵ=eps(),maxstep=128)     ##原函数、一阶导数、二阶导数、初始点；
    nstep = 0
    if abs(norm(g(x0...))) <= ϵ
        return x0,f(x0...)
    else
        while abs(norm(g(x0...))) > ϵ && nstep < maxstep
            println("step = ",nstep,"\t","x = ",x0,"\t","derivattive = ",g(x0...))
            nstep += 1
            x0 = x0 - inv(h(x0...))*g(x0...)  
        end
    end
    return x0,f(x0...)
end
```

```julia
function Newton(f,g,h,x0;
        ϵg=0.001,ϵx=0.001,ϵf=0.01,ϵ=eps(),maxstep=128)     ##原函数、一阶导数、二阶导数、初始点；
    i = 0
    if abs(norm(g(x0))) <= ϵ
        return x0,f(x0)
    else
        while abs(norm(g(x0))) > ϵ && i < maxstep
            println("step = ",i,"\t","x = ",x0,"\t","    f=",f(x0),"derivattive = ",g(x0))
            i += 1
            if det(h(x0))==0                    ##判断黑塞矩阵是否可逆(行列式为0不可逆)；
                println("ERROR : H Matrix irreversible!","\r","Can't use Newton! ")
            else
                x0 = x0 - inv(h(x0))*g(x0)  
            end
        end
        return x0,f(x0)
    end
end
```

```julia
function Levenberg(f,g,h,I,x0,μ=eps(),ϵ=0,maxIterration=128)
    x=x0
    xg=g(x...)
    xh=h(x...)
    α=0.01
    i=0
    if abs(norm(xg))<ϵ
        return x,f(x...)
    else
        while (abs(norm(xg))>ϵ && i<maxIterration)
            i+=1        
            α=0.01
            d=(inv(h(x...)+μ*I))*g(x...)
            x+=-α*d
            println("step = ",i,"  x = ",x,"  f=",f(x...),"  g = ",g(x...),"  h = ",h(x...),"  d=",d)    
        end
        return x,f(x...)
    end
end
```
