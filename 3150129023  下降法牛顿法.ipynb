梯度下降法

```julia
#第3题
#步长
function inexact_method(f,g,ϕ0,ϕd0,xk,dk;τ=0.5,ϵ=0.5,ζ=2)
    α=1
    dα=dk'*g((xk.+α.*dk)...)
    while dα<0
        α=ζ*α
        dα=dk'*g((xk.+α.*dk)...)
    end
    ϕα=f((xk.+α.*dk)...)
    while(ϕα>ϕ0+ϵ*α*ϕd0)
        α=τ*α
        ϕα=f((xk.+α.*dk)...)
    end
    return α 
end
function gradient_descent(f,g,x0;
        ϵg=0.0001,ϵx=0.0001,ϵf=0.0001,maxIteractions=128,verbose=true)
    x1=x0
    d=-g(x1...)
    f1=f(x1...)
    ng=norm(d)
    α=inexact_method(f,g,f1,-ng'*ng,x1,d)
    x2=x1+α.*d
    f2=f(x2...)
    nx=α*ng
    ni=1
    println("step:",ni,"   x:",x2,"   f:",f2,"  α=",α,"    g:",-d)
    while((ng>ϵg || nx>ϵx || abs(f2-f1)>ϵf )&& ni<maxIteractions )
        x1=x2
        d=-g(x1...)
        f1=f2
        ng=norm(d)
        α=inexact_method(f,g,f1,-ng'*ng,x1,d)
        x2=x1+α.*d
        f2=f(x2...)
        nx=α*ng                  
        ni+=1
        if verbose
            println("step:",ni,"   x:",x2,"   f:",f2,"    g:",-d)
            println("   d:" ,d,"  α:",α)
        end
    end
    if ni>=maxIteractions
        println("warning:Iteractions exceeds",maxIteractions)
    end
    return x2,f2
end
```

```julia
#第4题
gradient_descent(
    (x,y)->(1-x)^2+100(y-x^2)^2,
    (x,y)->[2*(x-1)+200*(y-x^2)*(-2*x),200*(y-x^2)],
    [-2,2]
)
```

牛顿法

```julia
#第3题 牛顿法
function newton_method(f,g,h,x0,ϵ=0,maxIterration=128)
    x=x0
    xg=g(x...)
    xh=h(x...)
    i=0
    if abs(norm(xg))<ϵ
        return x,f(x...)
    else
        while (abs(norm(xg))>ϵ && i<maxIterration)
            i+=1
            println(i,"\t","x=",x,"\tg=",xg,"\tH=",xh,"\t-H=",inv(xh),"\t-Hg",inv(xh)*xg,"\tf=",f(x...))
            x+=-inv(xh)*xg
            xg=g(x...)
            xh=h(x...)
        end
        return x,f(x...)
    end
end
```

```julia
#第4题 
newton_method(
    (x,y)->(1-x)^2+100(y-x^2)^2,
    (x,y)->[2*(x-1)+200*(y-x^2)*(-2*x) 200*(y-x^2)]',
    (x,y)->[2+800*x^2 -400*x;-400*x 200],
    [-2 2]'
)
```

```julia
#第3题 Levenberg-Marquardt 改进算法
function Levenberg(f,g,h,I,x0,μ=eps(),ϵ=0,maxIterration=128)
    x=x0
    xg=g(x...)
    xh=h(x...)
    α=0.1
    i=0
    if abs(norm(xg))<ϵ
        return x,f(x...)
    else
        while (abs(norm(xg))>ϵ && i<maxIterration)
            i+=1
            println(i,"\t","x=",x,"\tg=",xg,"\tH=",xh,"\t-H=",inv(xh),"\t-Hg",inv(xh)*xg,"\tf=",f(x...))
            α=0.1
            x+=-α*(inv(xh+μ*I))*xg
            xg=g(x...)
            xh=h(x...)
        end
        return x,f(x...)
    end
end
```

```julia
Levenberg(
    (x,y)->(1-x)^2+100(y-x^2)^2,
    (x,y)->[2*(x-1)+200*(y-x^2)*(-2*x) 200*(y-x^2)]',
    (x,y)->[2+800*x^2 -400*x;-400*x 200],
    [1 0;0 1],
    [-2 2]'
)
```
